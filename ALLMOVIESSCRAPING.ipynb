```python
import requests as rq
from bs4 import BeautifulSoup
```

```python
filename = "AllMovies.csv"
f = open(filename,"w")
headers = "Name,Year,Certificate,Length,Genre,Rating,Metascore,Votes,Gross\n"
f.write(headers)
```

```python
for i in range(1,101):
    my_url='https://www.imdb.com/list/ls057823854/?sort=list_order,asc&st_dt=&mode=detail&page='+str(i)
    pgresp=rq.get(my_url)
    soup= BeautifulSoup(pgresp.content,"html.parser")
    
    containers = soup.find_all('div',class_='lister-item mode-detail')
    
    for container in containers:
        Name = container.h3.a.string
        Yearf = container.h3.find("span",class_='lister-item-year')
        if Yearf is None or Yearf.string is None:
            Year = "NaN"
        else:
            Year = Yearf.string.replace("(","").replace(")","").replace(" ","").replace("I","").replace(",","")
    
        Certificatef = container.find("span",class_="certificate")
        if Certificatef is None:
            Certificate = "NaN";
        else:
            Certificate = Certificatef.string

        Lengthf = container.find("span",class_="runtime")
        if Lengthf is None:
            Length = "NaN"
        else: 
            Length = Lengthf.string.replace(" min","")
    
        Genref = container.find("span",class_="genre")
        if Genref is None:
            Genre = "NaN";
        else:
            Genre = Genref.string.replace(" ","").replace(",","|").replace("\n","")
        
        
        
        Ratingf = container.find(class_='ipl-rating-star__rating')
        if Ratingf is None:
            Rating = "NaN";
        else:
            Rating = Ratingf.string.replace(" ","")
        Metascoref =container.find("span",class_="metascore")
        if Metascoref is None:
            Metascore = "NaN";
        else:
            Metascore = Metascoref.string.replace(" ","")
        ppp = container.find_all('span',{"name":"nv"})
        if len(ppp)==0:
            Votes = "NaN"
            Gross = "NaN"
        elif len(ppp)<2:
            Votes = ppp[0]['data-value']
            Gross = "NaN"
        else:
            Votes = ppp[0]['data-value']
            Gross = ppp[1]['data-value'].replace(",","")
    
        #print(Name.replace(",","-") + "," + Year +","+Certificate +"," + Length.replace(" min","").replace("\n","") + ","+Genre.replace(" ","").replace(",","|").replace("\n","") + ","+Rating+","+Metascore+","+Votes+","+Gross + "\n")
        f.write(Name.replace(",","-") + "," + Year +","+Certificate +"," + Length.replace(" min","").replace("\n","") + ","+Genre.replace(" ","").replace(",","|").replace("\n","") + ","+Rating+","+Metascore+","+Votes+","+Gross + "\n")
    print (i)
```

```python
f.close()
```
